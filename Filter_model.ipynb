### We import our preprocessed data

```python
df = pd.read_csv("/Users/user/Documents/Spam_Detection/Data/data_clear.csv") 
```

```python
df=df.drop(columns='index',axis=1)
```

### We create our train and test dataframe

```python
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

### Most common words among spam analysis

```python
spam_emails = df_train[df_train["Category"] == 1]["Message"]
```

```python
word_list = []
for email in spam_emails:
    word_list.extend(Functions.clean_text(email))
```

```python
word_counts = Counter(word_list)
```

```python
spam_keywords = [word for word, count in word_counts.most_common(20)]
```

```python
print(spam_keywords)
```

### Now we test if our filter works

```python
y_pred = [Functions.is_spam(email,spam_keywords) for email in df_test['Message']]
y_true = [email for email in df_test['Category']]
```

```python
y_pred
```

```python
y_true
```

```python
conf_matrix = confusion_matrix(y_true, y_pred)
accuracy = accuracy_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
precision=precision_score(y_true, y_pred)

print("Confusion Matrix :\n", conf_matrix)
print(f"Accuracy : {accuracy:.2f}")
print(f"Precision : {precision:.2f}")
print(f"Recall : {recall:.2f}")
```

Our accuracy is pretty bad, only 30% of spam are detected but we have a very good recall 99%

### Now lets use an other type of filter

We now use an heuristic which is detailed in our Functions.py file

```python
y_pred = [Functions.is_spam_heuristic(email,spam_keywords) for email in df_test['Message']]
y_true = [email for email in df_test['Category']]
```

```python
conf_matrix = confusion_matrix(y_true, y_pred)
accuracy = accuracy_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
precision=precision_score(y_true, y_pred)

print("Confusion Matrix :\n", conf_matrix)
print(f"Accuracy : {accuracy:.2f}")
print(f"Precision : {precision:.2f}")
print(f"Recall : {recall:.2f}")
```

The accuracy is strongly improved but the recall is way under what we had with the much simpler filter
